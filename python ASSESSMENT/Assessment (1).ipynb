```python
# Q. 1
class ClinicAppointment:
    def __init__(self):
        self.doctors = {
            "Dr. Sharma": {},
            "Dr. Patel": {},
            "Dr. Mehta": {}
        }
        self.time_slots = ["10am", "11am", "12pm", "2pm", "3pm"]
        self.appointments = {}

    def book_appointment(self):
        name = input("Enter patient name: ")
        age = input("Enter patient age: ")
        mobile = input("Enter mobile number: ")

        if mobile in self.appointments:
            print(" You already have an appointment booked.")
            return

        print("\nAvailable Doctors:")
        for i, doc in enumerate(self.doctors.keys(), start=1):
            print(f"{i}. {doc}")

        choice = int(input("Choose doctor (1-3): "))
        doctor = list(self.doctors.keys())[choice - 1]

        print("\nAvailable Time Slots:")
        for i, slot in enumerate(self.time_slots, start=1):
            count = len(self.doctors[doctor].get(slot, []))
            status = "Available" if count < 3 else "Full"
            print(f"{i}. {slot} - {status}")

        slot_choice = int(input("Choose time slot (1-5): "))
        time_slot = self.time_slots[slot_choice - 1]

        # Check availability
        if len(self.doctors[doctor].get(time_slot, [])) >= 3:
            print("This slot is full. Please choose another one.")
            return

        # Book appointment
        self.doctors[doctor].setdefault(time_slot, []).append(mobile)
        self.appointments[mobile] = {
            "name": name,
            "age": age,
            "doctor": doctor,
            "time_slot": time_slot
        }
        print(f"\nAppointment confirmed for {name} with {doctor} at {time_slot}.")

    def view_or_cancel(self):
        mobile = input("Enter your mobile number: ")

        if mobile not in self.appointments:
            print("No appointment found with this number.")
            return

        appointment = self.appointments[mobile]
        print("\nAppointment Details:")
        print(f"Name: {appointment['name']}")
        print(f"Age: {appointment['age']}")
        print(f"Doctor: {appointment['doctor']}")
        print(f"Time Slot: {appointment['time_slot']}")

        choice = input("\nDo you want to cancel this appointment? (yes/no): ").lower()
        if choice == "yes":
            doctor = appointment["doctor"]
            time_slot = appointment["time_slot"]
            self.doctors[doctor][time_slot].remove(mobile)
            del self.appointments[mobile]
            print("Appointment cancelled successfully.")
        else:
            print(" Appointment kept as is.")

    def run(self):
        while True:
            print("\n===== Clinic Appointment System =====")
            print("1. Book Appointment")
            print("2. View/Cancel Appointment")
            print("3. Exit")

            choice = input("Enter your choice: ")

            if choice == "1":
                self.book_appointment()
            elif choice == "2":
                self.view_or_cancel()
            elif choice == "3":
                print("Thank you for using the system. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")


# Run the program
if __name__ == "__main__":
    clinic = ClinicAppointment()
    clinic.run()
```

```python
# Q.2
class SchoolManagement:
    def __init__(self):
        self.students = {}
        self.next_id = 1

    def new_admission(self):
        print("\n--- New Admission ---")
        name = input("Enter student name: ")
        age = int(input("Enter age: "))
        student_class = int(input("Enter class (1-12): "))
        guardian_mobile = input("Enter guardian mobile number: ")

        # Validate age
        if age < 5 or age > 18:
            print("Invalid age. Age must be between 5 and 18.")
            return

        # Validate mobile number
        if len(guardian_mobile) != 10 or not guardian_mobile.isdigit():
            print("Invalid mobile number. It must be 10 digits.")
            return

        # Assign unique student ID
        student_id = self.next_id
        self.next_id += 1

        # Store student details
        self.students[student_id] = {
            "Name": name,
            "Age": age,
            "Class": student_class,
            "GuardianMobile": guardian_mobile
        }

        print(f"Student admitted successfully. Assigned ID: {student_id}")

    def view_student(self):
        print("\n--- View Student Details ---")
        student_id = int(input("Enter student ID: "))

        if student_id in self.students:
            print("Student Details:")
            for key, value in self.students[student_id].items():
                print(key, ":", value)
        else:
            print("Student not found.")

    def update_student(self):
        print("\n--- Update Student Info ---")
        student_id = int(input("Enter student ID: "))

        if student_id in self.students:
            print("1. Update Mobile Number")
            print("2. Update Class")
            choice = input("Enter your choice: ")

            if choice == "1":
                new_mobile = input("Enter new mobile number: ")
                if len(new_mobile) == 10 and new_mobile.isdigit():
                    self.students[student_id]["GuardianMobile"] = new_mobile
                    print("Mobile number updated successfully.")
                else:
                    print("Invalid mobile number.")
            elif choice == "2":
                new_class = int(input("Enter new class (1-12): "))
                if 1 <= new_class <= 12:
                    self.students[student_id]["Class"] = new_class
                    print("Class updated successfully.")
                else:
                    print("Invalid class.")
            else:
                print("Invalid choice.")
        else:
            print("Student not found.")

    def remove_student(self):
        print("\n--- Remove Student Record ---")
        student_id = int(input("Enter student ID: "))

        if student_id in self.students:
            del self.students[student_id]
            print("Student record removed successfully.")
        else:
            print("Student not found.")

    def run(self):
        while True:
            print("\n===== School Management System =====")
            print("1. New Admission")
            print("2. View Student Details")
            print("3. Update Student Info")
            print("4. Remove Student Record")
            print("5. Exit")
            choice = input("Enter your choice: ")

            if choice == "1":
                self.new_admission()
            elif choice == "2":
                self.view_student()
            elif choice == "3":
                self.update_student()
            elif choice == "4":
                self.remove_student()
            elif choice == "5":
                print("Exiting system. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")


# Run the program
school = SchoolManagement()
school.run()
```

```python
# Q.3
class BusReservation:
    def __init__(self):
        self.routes = {
            "Mumbai to Pune": 500,
            "Delhi to Jaipur": 600,
            "Bangalore to Mysore": 450,
            "Chennai to Coimbatore": 550,
            "Kolkata to Durgapur": 400
        }
        self.tickets = {}
        self.next_ticket_id = 1
        self.seats_booked = {route: 0 for route in self.routes}

    # Show Available Routes (with numbers)
    def show_routes(self):
        print("\nAvailable Routes and Prices:")
        for i, (route, price) in enumerate(self.routes.items(), start=1):
            print(f"{i}. {route} - Rs.{price}")

    # Book Ticket
    def book_ticket(self):
        print("\nBook Ticket")
        name = input("Enter passenger name: ")
        age = int(input("Enter age: "))
        mobile = input("Enter mobile number: ")

        if len(mobile) != 10 or not mobile.isdigit():
            print("Invalid mobile number. Must be 10 digits.")
            return

        # Show numbered routes
        self.show_routes()

        try:
            route_choice = int(input("Select a route by number: "))
            routes_list = list(self.routes.keys())

            if 1 <= route_choice <= len(routes_list):
                route = routes_list[route_choice - 1]
            else:
                print("Invalid route selection.")
                return

        except ValueError:
            print("Please enter a valid number.")
            return

        if self.seats_booked[route] >= 40:
            print("No seats available for this route.")
            return

        seat_number = self.seats_booked[route] + 1
        self.seats_booked[route] += 1
        ticket_id = self.next_ticket_id
        self.next_ticket_id += 1

        self.tickets[ticket_id] = {
            "name": name,
            "age": age,
            "mobile": mobile,
            "route": route,
            "seat_number": seat_number,
            "price": self.routes[route]
        }

        print(f"\nTicket booked successfully!")
        print(f"Your Ticket ID is {ticket_id}")
        print(f"Route: {route}")
        print(f"Seat Number: {seat_number}")
        print(f"Ticket Price: Rs.{self.routes[route]}")

    # View Ticket
    def view_ticket(self):
        print("\nView Ticket")
        try:
            ticket_id = int(input("Enter Ticket ID: "))
        except ValueError:
            print("Invalid input. Enter a number.")
            return

        if ticket_id in self.tickets:
            t = self.tickets[ticket_id]
            print("\nTicket Details:")
            print(f"Passenger Name: {t['name']}")
            print(f"Age: {t['age']}")
            print(f"Mobile: {t['mobile']}")
            print(f"Route: {t['route']}")
            print(f"Seat Number: {t['seat_number']}")
            print(f"Ticket Price: Rs.{t['price']}")
        else:
            print("Ticket not found.")

    # Cancel Ticket
    def cancel_ticket(self):
        print("\nCancel Ticket")
        try:
            ticket_id = int(input("Enter Ticket ID to cancel: "))
        except ValueError:
            print("Invalid input. Enter a number.")
            return

        if ticket_id in self.tickets:
            route = self.tickets[ticket_id]["route"]
            del self.tickets[ticket_id]
            self.seats_booked[route] -= 1
            print("Ticket cancelled successfully.")
        else:
            print("Ticket not found.")

    # Exit System
    def exit_system(self):
        print("\nExiting Bus Reservation System. Thank you!")
        exit()


# Main Program
if __name__ == "__main__":
    bus = BusReservation()

    while True:
        print("\n=== Bus Reservation System ===")
        print("1. Show Available Routes")
        print("2. Book Ticket")
        print("3. View Ticket")
        print("4. Cancel Ticket")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            bus.show_routes()
        elif choice == "2":
            bus.book_ticket()
        elif choice == "3":
            bus.view_ticket()
        elif choice == "4":
            bus.cancel_ticket()
        elif choice == "5":
            bus.exit_system()
        else:
            print("Invalid choice. Please select between 1 and 5.")
```
