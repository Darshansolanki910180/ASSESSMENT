```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/historical_automobile_sales.csv',encoding='latin-1')
df.reset_index()
df
```

```python
df1 = df.groupby('Year')[['Automobile_Sales']].sum().reset_index()
df1
```

```python
#  Q 1: Develop a line chart using the functionality of pandas to show how automobile
#  sales fluctuate from year to year.

plt.figure(figsize=(10,5))
plt.plot(df1['Year'],df1['Automobile_Sales'],marker='o',color='blue')
plt.title("Year wise Sales")
plt.xlabel("Year")
plt.ylabel("Automobile_Sales")
plt.show()
```

```python
# Q 2: Plot different lines for categories of vehicle type and analyze the trend
# to answer the question Is there a noticeable difference in sales trends
# between different vehicle types during recession periods?

grouped = df.groupby(["Year", "Vehicle_Type"])["Automobile_Sales"].sum().reset_index()
grouped

```

```python
pivot_df = grouped.pivot(index="Year", columns="Vehicle_Type", values="Automobile_Sales")


pivot_df.plot(figsize=(12,6), marker='o')

plt.title("Year-wise Automobile Sales by Vehicle Type")
plt.xlabel("Year")
plt.ylabel("Automobile_Sales")
plt.grid(True)
plt.show()
```

```python
# Q 3: Use the functionality of Seaborn Library to create a visualization to compare
# the sales trend per vehicle type for a recession period with a non- recession
# period.

sales_comparison = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()


sales_comparison.rename(columns={'Automobile_Sales': 'Mean_Sales'}, inplace=True)
sales_comparison['Recession_Status'] = sales_comparison['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

plt.figure(figsize=(10, 5))
sns.barplot(
    x='Vehicle_Type',
    y='Mean_Sales',
    hue='Recession_Status',
    data=sales_comparison,
    palette=['skyblue', 'salmon'],
    errorbar=None
)


plt.title('Mean Automobile Sales Comparison by Vehicle Type: Recession vs. Non-Recession', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Mean Automobile Sales (in thousands)', fontsize=12)


plt.xticks(rotation=45, ha='right')

plt.legend(title='Period', loc='upper right')
plt.tight_layout()


plt.savefig('sales_trend_comparison_bar_chart.png')
```

```python

# Q 4: Now you want to compare the sales of different vehicle types
# during a recession and a non-recession period.

sales_comparison = df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

pivot_table = sales_comparison.pivot(index='Vehicle_Type', columns='Recession', values='Automobile_Sales')

pivot_table.columns = ['Non-Recession', 'Recession']

plt.figure(figsize=(10, 5))

pivot_table.plot(kind='bar', ax=plt.gca())
plt.title('Average Automobile Sales by Vehicle Type: Recession vs. Non-Recession', fontsize=16)
plt.xlabel('Vehicle Type', fontsize=12)
plt.ylabel('Average Sales (Units)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Period')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()



print("\nAverage Automobile Sales Comparison Table (Units):")
print(pivot_table)
```
